True Positive: 14 , False Positive: 2, False Negative: 1, True Negative: 3


```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

actual_result = np.array(['Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'No Disease','No Disease', 'No Disease', 'No Disease', 'No Disease' ])
predicted_result = np.array(['Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease', 'Disease','Disease', 'No Disease', 'Disease', 'Disease', 'No Disease', 'No Disease', 'No Disease'  ])

conf_matx = confusion_matrix(actual_result,predicted_result)

sns.heatmap(conf_matx, annot= True, fmt= 'g',
            xticklabels= ['Disease', 'No Disease'],
            yticklabels= ['Disease', 'No Disease'])
plt.xlabel('Predicted', fontsize=14)
plt.ylabel('Actual', fontsize=14)

plt.show()

accuracy = accuracy_score(actual_result, predicted_result)
print("Accuarcy:", accuracy)

precision = precision_score(actual_result, predicted_result, average="weighted")
print("Precision:", precision)

recall = recall_score(actual_result, predicted_result, average="weighted")
print("Recall:", recall)

f1score = f1_score(actual_result, predicted_result, average="weighted")
print("F1 Score:",f1score)
```
